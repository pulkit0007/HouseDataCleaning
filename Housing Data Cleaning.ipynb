```python
#Calculate Percentage of Mising values for every column
per = (data.isnull().sum()/len(data.index))*100
print(per)
```

```python
#Droping Columns with per value > 33
data = data.drop('BuildingArea',axis=1)
data = data.drop('YearBuilt',axis=1)
data = data.drop('CouncilArea',axis=1)
```

```python
data.shape
```

```python
data = data[~np.isnan(data['Price'])]  # ~ Negates. isnan will remove all false values so we negate
```

```python
data.shape
```

```python
data = data[~np.isnan(data['Landsize'])]
```

```python
data.shape
```

```python
per = (data.isnull().sum()/len(data.index))*100
print(per)
```

```python
data[['Lattitude','Longtitude']].describe()
```

```python
data['Lattitude'] = data['Lattitude'].fillna(data['Lattitude'].mean())
data['Longtitude'].fillna(data['Longtitude'].mean(), inplace=True)  #Either store it in a variable or use Inplace=True
```

```python
data['Bathroom'].value_counts()  # mode()[0]=1.0
```

```python
data['Car'].value_counts()  # mode()[0] = 2.0
```

```python
data['Bathroom'].fillna(data['Bathroom'].mode()[0], inplace = True)
data['Car'].fillna(data['Car'].mode()[0], inplace = True)
```
